#### BWT vs AGE

```python
model1 = smf.logit(formula="BWT1 ~ AGE", data=df1
                       ).fit()
```

```python
model1.summary()
```

```python
model1.summary2()
```

```python
# crude odds ratio confidence interval for age
c1 = model1.conf_int(alpha=0.05)
np.exp(c1)
```

```python
# crude odds ratio for age
np.exp(0.0512)
```

For an increase of 1 year in age of mother there is 5% more likelihood of Low Birth Weight in the mother.

#### BWT vs LWT

```python
model2 = smf.logit(formula="BWT1 ~ LWT", data=df1
                       ).fit()
print(model2.summary())
print(model2.summary2())

# crude odds ratio confidence interval for LWT
c2 = model2.conf_int(alpha=0.05)
np.exp(c2)

# crude odds ratio for LWT
print(np.round(np.exp(0.0141),2))
```

for increase of 1 pound of weight in last menstrual cycle there is 1% more likelihood of low birth weight in the mother

```python
model3 = smf.logit(formula="BWT1 ~ RACE", data=df1
                       ).fit()
print(model3.summary())
print(model3.summary2())

# crude odds ratio confidence interval for LWT
c3 = model3.conf_int(alpha=0.05)
np.exp(c3)

# crude odds ratio for LWT
print(np.round(np.exp(-0.8448),2))
print(np.round(np.exp(-0.6362),2))
```

by default Python takes the least code category as reference category so we have values of race2 and race3 relative to race1

race 1 -> OR =1 (reference) [no confidence interval]

race 2 -> OR interval

A mother belonging to race 2 has 53% less likelihood of having low birth weight in comparison to one belonging to race 1. Similarly, a mother belonging to race 3 has 43% less likelihood of having low birth weight in comparison to one belonging to race 1
