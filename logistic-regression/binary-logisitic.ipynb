### Logistic Regression Analysis

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
```

```python
df = pd.read_csv("LOWBWT.csv",)
df.head()
```

#### Variables

* lwt : weight in last menstrual cycle in pounds

* ftv : visit to the physician

* ptl : previous pre termed labour

* smoke : smoking during

* ht : hypertension during preganancy

* race

    * 1=white
    * 2=black
    * 3=others

* ui : uterine irritability

* bwt : birth weight in grams

* categorize birth weight to :

    * low bwt <2500gmhighÂ 
    * bwt >=2500gm

```python
df1 = df
```

```python
for i in range(len(df1)):
    if df1.loc[i,'BWT'] >= 2500:
        df1.loc[i,'BWT1'] = 1
    else:
        df1.loc[i,'BWT1'] = 0
```

```python
df1['BWT1'] = df1['BWT1'].astype('int')
```

```python
df1['BWT1'] = df1['BWT1'].astype('str')
```

```python
df1.info()
```

```python
df1.RACE = df1.RACE.astype('str')
df1.HT = df1.HT.astype('str')
df1.SMOKE = df1.SMOKE.astype('str')
df1.UI = df1.UI.astype('str')
df1.BWT = df1.BWT.astype('str')
```

```python
df1.info()
```

```python
import statsmodels.api as sm
df1.drop(["BWT"],axis=1,inplace=True)
df1
```

```python
df1.BWT1 = df1.BWT1.astype('int')
```

#### BWT vs AGE

```python
model1 = smf.logit(formula="BWT1 ~ AGE", data=df1
                       ).fit()
```

```python
model1.summary()
```

```python
model1.summary2()
```

```python
# crude odds ratio confidence interval for age
c1 = model1.conf_int(alpha=0.05)
np.exp(c1)
```

```python
# crude odds ratio for age
np.exp(0.0512)
```

For an increase of 1 year in age of mother there is 5% more likelihood of Low Birth Weight in the mother.

#### BWT vs LWT

```python
model2 = smf.logit(formula="BWT1 ~ LWT", data=df1
                       ).fit()
print(model2.summary())
print(model2.summary2())

# crude odds ratio confidence interval for LWT
c2 = model2.conf_int(alpha=0.05)
np.exp(c2)

# crude odds ratio for LWT
print(np.round(np.exp(0.0141),2))
```

for increase of 1 pound of weight in last menstrual cycle there is 1% more likelihood of low birth weight in the mother

```python
model3 = smf.logit(formula="BWT1 ~ RACE", data=df1
                       ).fit()
print(model3.summary())
print(model3.summary2())

# crude odds ratio confidence interval for LWT
c3 = model3.conf_int(alpha=0.05)
np.exp(c3)

# crude odds ratio for LWT
print(np.round(np.exp(-0.8448),2))
print(np.round(np.exp(-0.6362),2))
```

by default Python takes the least code category as reference category so we have values of race2 and race3 relative to race1

race 1 -> OR =1 (reference) [no confidence interval]

race 2 -> OR interval

race 3 -> OR interval

A mother belonging to race 2 has 53% less likelihood of having low birth weight in comparison to one belonging to race 1. Similarly, a mother belonging to race 3 has 43% less likelihood of having low birth weight in comparison to one belonging to race 1

#### BWT vs SMOKE

```python
model4 = smf.logit(formula="BWT1 ~ SMOKE", data=df1
                       ).fit()
print(model4.summary())
print(model4.summary2())

# crude odds ratio confidence interval for LWT
c4 = model4.conf_int(alpha=0.05)
np.exp(c4)

# crude odds ratio for LWT
print(np.round(np.exp(-0.7041),2))
```

there is 50% less likelihood of low birth weight in mothers who smoked during menstrual cycle compared to mothers who did not smoke.

### BWT vs PTL

```python
model5 = smf.logit(formula="BWT1 ~ PTL", data=df1).fit()
print(model5.summary())
print(model5.summary2())

# crude odds ratio confidence interval for LWT
c5 = model5.conf_int(alpha=0.05)
np.exp(c5)

# crude odds ratio for LWT
print(np.round(np.exp(-0.8081),2))
```

for mothers with PTL (greater than 0 number of days) there is 45% less likelihood of low birth weight in mothers compared to mothers with no pretermed labour.

### BWT vs HT

```python
model6 = smf.logit(formula="BWT1 ~ HT", data=df1).fit()
print(model6.summary())
print(model6.summary2())

# crude odds ratio confidence interval for LWT
c6 = model6.conf_int(alpha=0.05)
np.exp(c6)

# crude odds ratio for LWT
print(np.round(np.exp(-1.2135),2))
```

mother with hypertension during pregnancy have 70% less likelihood of low birth weight compared to mothers without hypertension

### BWT vs UI

```python
model7 = smf.logit(formula="BWT1 ~ UI", data=df1).fit()
print(model7.summary())
print(model7.summary2())

# crude odds ratio confidence interval for LWT
c7 = model7.conf_int(alpha=0.05)
np.exp(c7)

# crude odds ratio for LWT
print(np.round(np.exp(-0.9469),2))
```

for mothers who have had uterine irritability have 61 % less likelihood of low birth weight compared to those who did not have

### BWT vs FTV

```python
model8 = smf.logit(formula="BWT1 ~ FTV", data=df1).fit()
print(model8.summary())
print(model8.summary2())

# crude odds ratio confidence interval for LWT
c8 = model8.conf_int(alpha=0.05)
np.exp(c8)

# crude odds ratio for LWT
print(np.round(np.exp(0.1351),2))
```

mothers who visited physicians for have 14% more likelihood of low birth weight relative to mothers who didn't visit them.

### Multiple Logistic Regression

```python
model9 = smf.logit(formula="BWT1 ~ AGE+FTV+RACE+SMOKE+PTL+HT+UI", data=df1).fit()
print(model9.summary())
print(model9.summary2())

# crude odds ratio confidence interval for LWT
c9 = model9.conf_int(alpha=0.05)
np.exp(c9)

# crude odds ratio for LWT
print(np.round(np.exp(model9.params),2))
```
